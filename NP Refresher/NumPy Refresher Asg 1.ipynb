In the following code cell you will be provided with several NumPy arrays. Use those arrays to complete the following tasks:<br>

```python
a = np.array([[2, 2, 2], [3, 3, 3]])
b = np.array([[1, 1, 1], [4, 4, 4]])
c = np.array([[5, 5, 5], [6, 6, 6], [7, 7, 7], [8, 8, 8]])

print("Array a:")
print("\n", a)
print("\n", "Array b:")
print("\n", b)
print("\n", "Array c:")
print("\n", c)
```

# SOLUTION PROBLEM #9

```python
d = np.vstack([b,a])
d
```

```python
e = np.vstack([c,d])
e
```
